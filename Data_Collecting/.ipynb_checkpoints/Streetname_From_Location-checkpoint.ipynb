```python
import pandas as pd
from geopy.geocoders import Nominatim
from shapely import Point, wkt
import numpy as np
import os
import geopandas as gpd
```

```python
current_dir = os.getcwd()

# Construct the path to the 'data' directory relative to the current directory
data_dir = os.path.join(current_dir, '..', 'data')

# Access a specific file
map12_df = pd.read_csv(os.path.join(data_dir, 'Secondary_Primary_Streets.csv'))
map3r_df = pd.read_csv(os.path.join(data_dir, 'Tertiary_Residential_Streets.csv'))
```

```python
map_df = pd.concat([map12_df, map3r_df], ignore_index = True)
map_df
```

```python
problem_df = map_df[map_df["name"].isnull()].copy().reset_index(drop=True)
problem_df
```

```python
problem_df['geometry'] = problem_df['geometry'].apply(wkt.loads)
gdf = gpd.GeoDataFrame(problem_df, crs='epsg:4326')
gdf.plot()
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from shapely import Point
import numpy as np
import os
```

```python
current_dir = os.getcwd()

# Construct the path to the 'data' directory relative to the current directory
data_dir = os.path.join(current_dir, '..', 'data')

# Access a specific file
bike_df = pd.read_csv(os.path.join(data_dir, 'Infravelo_Clean.csv'))
bike_df.head()
```

```python
# test
address = "Adorfer Straße, Neukölln"

def get_lat_lon(address):
    address = address + ", Berlin"
    geolocator = Nominatim(user_agent="my_geocoding_app")
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        print(f"Problematic adress: {address}")
        return None, None

# Example usage
address = "Adorfer Straße 8"
lat, lon = get_lat_lon(address)
print(f"Latitude: {lat}, Longitude: {lon}")
```

```python
test_df = bike_df.copy()
```

```python
test_df = test_df[test_df["function"] == "parking"]
test_df
```

```python
test_df["title"].value_counts()
```

```python
latlon = []

for address in test_df["title"].tolist():
    lat, lon = get_lat_lon(address)
    latlon.append([lat, lon])
```

```python
latlon
```

```python
test_df["title"].tolist()
```

```python
# transform to point format
shapely_loc = []
for i in range(len(latlon)):
    loc = latlon[i]
    if loc != [None, None]:
        upd_loc = Point(loc[0],loc[1])
        shapely_loc.append(upd_loc)
    else:
        shapely_loc.append(np.nan)
shapely_loc
```

```python
bike_df.loc[bike_df["title"].notnull(), "location"] = shapely_loc
```

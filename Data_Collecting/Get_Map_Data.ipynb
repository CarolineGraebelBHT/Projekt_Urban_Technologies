```python
# map builder
import pandas as pd
import geopandas as gpd
import os
```

```python
current_dir = os.getcwd()

data_dir = os.path.join(current_dir, '..', 'data')

gdf1 = gpd.read_file(os.path.join(data_dir, "primary_secondary.geojson"), encoding='utf-8')
gdf2 = gpd.read_file(os.path.join(data_dir, "residential_tertiary.geojson"), encoding='utf-8')
```

```python
#gdf1 = gpd.read_file("primary_secondary.geojson", encoding='utf-8')
gdf1.plot()
```

```python
#gdf2 = gpd.read_file("residential_tertiary.geojson", encoding='utf-8')
gdf2.plot()
```

```python
# choose relevant variables for further processing
gdf1_sub = gdf1.loc[:,["highway", "lanes", "lit", "maxspeed", "name", "oneway", "ref", "smoothness", "geometry"]]
gdf1_sub.head()
```

```python
gdf2_sub = gdf2.loc[:,["highway", "lanes", "lit", "maxspeed", "name", "oneway", "ref", "smoothness", "geometry"]]
gdf2_sub.head()
```

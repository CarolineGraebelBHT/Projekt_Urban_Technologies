```python
# building a map
import pandas as pd
import os
import geopandas as gpd
from shapely import wkt
import geopandas as gpd
import requests
```

```python
# import data
current_dir = os.getcwd()

# Construct the path to the 'data' directory relative to the current directory
data_dir = os.path.join(current_dir, '..', 'data')

# Access a specific file
map12_df = pd.read_csv(os.path.join(data_dir, 'Secondary_Primary_Streets.csv'))
map3r_df = pd.read_csv(os.path.join(data_dir, 'Tertiary_Residential_Streets.csv'))
bike_df = pd.read_csv(os.path.join(data_dir, 'Infravelo+Location.csv'))
```

```python
# merge maps
map_df = pd.concat([map12_df, map3r_df], ignore_index = True)
map_df
```

```python
map_df = map_df.rename(columns = {"name": "streetname"})
map_df
```

```python
# show data points where the streetname is missing
problem_df = map_df[map_df["streetname"].isnull()].copy().reset_index(drop=True)
problem_df
```

```python
# drop data point where streetname is missing
map_df = map_df[map_df["streetname"].notnull()]
map_df
```

```python
bike_df = bike_df.rename(columns = {"title": "streetname"})
bike_df.head()
```

```python
bike_df[bike_df["location"].isnull()]
```

```python
# what kind of streets don't have a location info?
bike_df[bike_df["location"].isnull()]["streetname"].value_counts()
```

```python
map_bike_df = map_df.merge(bike_df, on = "streetname", how = "outer")
map_bike_df
```

```python
test = map_df[map_df["streetname"] == "Alt-Moabit"]
test
```

```python
def process_linestring(linestring):
    cleaned_string = linestring[12:-1]
    loc_list = cleaned_string.split(',')
    coordinates = []
    for pair in loc_list:
        lon, lat = map(float, pair.split())
        coordinates.append((lon, lat))
    return coordinates

geo_list = test["geometry"].to_list()

loc_list = []
for i in range(len(geo_list)):
    transformed_geo = process_linestring(geo_list[i])
    loc_list.append(transformed_geo)

test["altgeom"] = loc_list

```

```python
import pandas as pd
import numpy as np
import os
```

```python
# import data
current_dir = os.getcwd()

# Construct the path to the 'data' directory relative to the current directory
data_dir = os.path.join(current_dir, '..', 'data')

# Access a specific file
df = pd.read_csv(os.path.join(data_dir, 'infravelo_projects_raw.csv'))
```

```python
# filter rows only for finished projects
df = df[df["status"] == "Abgeschlossen"]
```

```python
# only take columns that are useful for a route planner project
df = df.loc[:,["id", "title", "subtitle", "link", "district", "type", "name", "value", "categories"]]
df
```

```python
# 1) Fix type column so it's an object descriptor
df["type"].value_counts()
```

```python
df[df["type"] == "Farblich beschichtet"]["subtitle"]
# actual type is hidden here
# every item listed here has gotten a green bike lane
```

```python
df_clean = df.copy()

df_clean.loc[df_clean["type"] == "Farblich beschichtet","type"] = "Grüner Seitenstreifen"
```

```python
df_clean["type"].value_counts()
```

```python
df[df["type"] == "Radwegbeschichtung /-durchfärbung"]["subtitle"]
```

```python
df_clean.loc[df_clean["type"] == "Radwegbeschichtung /-durchfärbung","type"] = "Grüner Seitenstreifen"
```

```python
df[df["type"] == "Elektrokleinstfahrzeuge"]
# after researching this, these are parking spots for small motorbikes, not really relevant for a biking route planner
```

```python
index_elektro = df_clean[df_clean["type"] == "Elektrokleinstfahrzeuge"].index
df_clean.drop(index_elektro, inplace = True)

index_verkehrsübungsplatz = df_clean[df_clean["type"] == "Verkehrsübungsplatz"].index
df_clean.drop(index_verkehrsübungsplatz, inplace = True)

index_verkehrsübungsplatz = df_clean[df_clean["type"] == "Verkehrsübungsplatz"].index
df_clean.drop(index_verkehrsübungsplatz, inplace = True)
```

```python
df_clean["type"].value_counts()
```

```python
# make a new variable "function" that holds general info about whether this object is some kind of bike parking or a road
parking = ["Kinderfahrradbügel","Anlehnbügel", "Radfahrstreifen", "Lastenfahrradbügel", "Vorderradhalter Rahmen/Gabel", "Fahrradbox"]

df_clean['function'] = df_clean['type'].apply(lambda x: 'parking' if x in parking else 'road')
df_clean.head()
```

```python
df_clean = df_clean.drop("categories", axis = 1)
df_clean = df_clean.drop("subtitle", axis = 1)
```

```python
df_clean.to_csv("Infravelo_Clean.csv", index = False)
```
